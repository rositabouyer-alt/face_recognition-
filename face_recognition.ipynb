```python
cap = cv2.VideoCapture(0)

while True:
    ret, img = cap.read()
    if not ret:
        break

    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    locate = fr.face_locations(img2)

    for (y1, x2, y2, x1) in locate:
        encodes = fr.face_encodings(img2, [(y1, x2, y2, x1)])
        if len(encodes) == 0:
            continue

        encode = encodes[0]
        result = fr.compare_faces([sabte_sorat, sabte_sorat2], encode)

        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 6)

        if result[0]:
            name = "Rozita"
        elif result[1]:
            name = "XXX"
        else:
            name = "***"

        cv2.putText(img, name, (x1 + 6, y2 + 20), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    cv2.imshow('image', img)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
